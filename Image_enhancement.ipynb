```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread("/content/Image 1.jpeg", cv2.IMREAD_GRAYSCALE)

plt.figure(figsize=(10, 5))
plt.subplot(1, 4, 1)
plt.title("Original Image")
plt.imshow(image, cmap='gray')
plt.axis('off')

def gamma_correction(img, gamma):
    gamma_corrected = np.array(255 * (img / 255) ** gamma, dtype='uint8')
    return gamma_corrected

gamma_image = gamma_correction(image, gamma=0.5)

plt.subplot(1, 4, 2)
plt.title("Gamma Correction")
plt.imshow(gamma_image, cmap='gray')
plt.axis('off')

def histogram_equalization(img):
    hist_eq_image = cv2.equalizeHist(img)
    return hist_eq_image

hist_eq_image = histogram_equalization(image)

plt.subplot(1, 4, 3)
plt.title("Histogram Equalization")
plt.imshow(hist_eq_image, cmap='gray')
plt.axis('off')

def log_transformation(img):
    c = 255 / (np.log(1 + np.max(img)))
    log_image = c * (np.log(1 + img))
    log_image = np.array(log_image, dtype=np.uint8)
    return log_image

log_image = log_transformation(image)

plt.subplot(1, 4, 4)
plt.title("Log Transformation")
plt.imshow(log_image, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```
